Pandas Queries & Data Manipulation Tasks

```python
import pandas as pd

df = pd.DataFrame({
    "festival": ["Carnival", "Oktoberfest", "La Tomatina", "Rio New Year", "Harbin Ice Festival", "Lantern Festival", "Cherry Blossom Festival", "Dia de los Muertos"],
    "country": ["Brazil", "Germany", "Spain", "Brazil", "China", "China", "Japan", "Mexico"],
    "month": ["February", "September", "August", "January", "January", "February", "April", "November"],
    "theme": ["Parade", "Beer Celebration", "Tomato Fight", "Fireworks", "Ice Sculptures", "Lanterns", "Nature", "Tradition"],
    "popularity_score": [98, 95, 88, 92, 90, 85, 89, 91]
})
```

```python
df
```

1. Select all festivals in China.

```python
df[df["country"]=="China"]
```

2 Sort festivals by popularity (descending)

```python
df.sort_values("popularity_score",ascending=False)
```

3 Show festival and theme columns only.

```python
df.loc[:,["festival","theme"]]
```

4 Filter festivals in Brazil or Japan.

```python
df[df.country.isin(['Brazil','Japan'])]
```

5. Find the most popular festival.

```python
df.sort_values('popularity_score',ascending=False).head(1)
```

6. Count how many festivals fall under each "theme".

```python
df.value_counts('theme')
```

7.Check festivals where the theme contains "Lantern".

```python
df[df.theme.str.contains('Lantern')]
```

8. Select rows 2 to 5 using iloc

```python
df.iloc[2:6]
```

9. Select festival names where popularity > 90.

```python
df[df["popularity_score"]>90]
```

10. Get unique countries.

```python
df.country.unique()
```
