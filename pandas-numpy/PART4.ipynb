Creating a DataFrame

```python
# Sample data
import pandas as pd

df = pd.DataFrame({
    "country": ["Italy", "US", "France", "Italy", "Australia", "US", "Spain", "Italy", "France", "US"],
    "province": ["Tuscany", "California", "Bordeaux", "Sicily", "Barossa", "Oregon", "Rioja", "Piedmont", "Champagne", "Washington"],
    "title": ["Vino Nobile", "Napa Valley Red Blend", "Chateau du Test", "Nero d‚ÄôAvola",
              "Barossa Shiraz", "Willamette Pinot Noir", "Reserva Tempranillo",
              "Barolo", "Brut Blanc de Blancs", "Columbia Valley Chardonnay"],
    "points": [88, 92, 95, 90, 91, 89, 93, 97, 96, 87],
    "price": [15, 45, 80, 18, 35, 28, 22, 120, 16, 18]  
})

df

```

Single column as a Series

```python
df["title"] # Accessing the 'title' column
```

Selecting Multiple columns

```python
df[["country","province","title"]] # Accessing multiple columns
```

Renaming Columns

```python
df.rename(columns={'country':'nation'}, inplace=True) # Renaming a column
df
```

Removing Columns

```python
df.drop("nation",axis=1, inplace=True) # Dropping a column
```

Adding Columns

```python
df['user-comments']=["Good wine","Excellent","Superb","Very good","Great value","Nice taste","Highly recommended","Outstanding","Crisp and refreshing","Decent"]
df
```

. Sorting Values

```python
df.sort_values("price")# Sorting by price in ascending order
```

```python
df.sort_values("price",ascending=False)# Sorting by price in descending order. ascending=False tells pandas to sort in descending order.
```

Multiple columns

```python
df.sort_values(["price","points"],ascending=False) # Sorting by price and points in descending order.
```

Filtering by condition

single condition

```python
df[df["points"]>80] # Filtering rows where points are greater than 80
```

Multiple conditions -AND

```python
df[(df["user-comments"]=="Excellent") & (df["user-comments"]=="Outstanding")] # This will return an empty DataFrame since no entry can be both "Excellent" and "Outstanding" at the same time.
```

OR

```python
df[(df["user-comments"]=="Excellent") | (df["user-comments"]=="Outstanding")]  # Filtering rows where user-comments is either "Excellent" or "Outstanding"
```
