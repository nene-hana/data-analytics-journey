Creating a DataFrame

```python
# Sample data
import pandas as pd

reviews = pd.DataFrame({
    "country": ["Italy", "US", "France", "Italy", "Australia", "US", "Spain", "Italy", "France", "US"],
    "province": ["Tuscany", "California", "Bordeaux", "Sicily", "Barossa", "Oregon", "Rioja", "Piedmont", "Champagne", "Washington"],
    "title": ["Vino Nobile", "Napa Valley Red Blend", "Chateau du Test", "Nero d‚ÄôAvola",
              "Barossa Shiraz", "Willamette Pinot Noir", "Reserva Tempranillo",
              "Barolo", "Brut Blanc de Blancs", "Columbia Valley Chardonnay"],
    "points": [88, 92, 95, 90, 91, 89, 93, 97, 96, 87],
    "price": [15, 45, 80, 18, 35, 28, 22, 120, 16, 18]  
})

reviews

```

1. **Attribute Access & Basic Indexing**

How do you access the country column using attribute access and indexing access?

Why is using df.country sometimes unsafe? Give an example.

Using `df.column` is sometimes unsafe because it can fail if the column name has spaces, special characters, or conflicts with a DataFrame method.
for example if the column name is 'republican countries' you cannot use reviews.republican countries instead you have to use reviews['republican countries']

**2. Understanding loc**

Using loc, select the row where the index label is 9

```python
reviews.loc[9]
```

Select the rows with index labels 0, 5, and 9, and only the columns country and points.

```python
reviews.loc[[0,5,9],["country","points"]]
```

**3 Understanding iloc**

Using iloc, select the first row.

```python
reviews.iloc[0] #first row accessing
```

Select rows at positions 0, 2, and 4, and columns at positions 1 and 3.

```python
reviews.iloc[[0,2,4],[1,3]] #specific rows and columns
```

**4. Conditional Selection**

Select all rows where points is greater than 90.

```python
reviews[reviews["points"] > 90]
```

Select all rows where country is either France or Italy

```python
reviews[(reviews["country"]=="France") | (reviews["country"]=="Italy")]
```
