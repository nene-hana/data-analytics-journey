# Level 5 ‚Äî Adding Columns

Add a new column size_category based on avg_length_cm: 1000 ‚Üí "giant" 100 ‚Üí "large" ,50 ‚Üí "medium" else ‚Üí "small

```python
import numpy as np
conditions= [
    animals["length"]>1000,
    animals["length"]>100,
    animals["length"]>50
]
choices=["giant","large","medium"]
animals["size_category"]=np.select(conditions,choices,default="small")
animals
```

or

```python
def categorize_length(length):
    if length > 1000:
        return "giant"
    elif length > 100:
        return "large"
    elif length > 50:
        return "medium"
    else:
        return "small"

animals["size_category"] = animals["length"].apply(categorize_length)
animals
```

Add a new column is_heavy: True if avg_weight_kg > 500 False otherwise

```python
def heav(weight):
    if weight>500:
        return "heavy"
    else:
        return "false"
    
animals["is_heavy"]=animals["weight"].apply(heav)
animals
```
